## Tabla de todas las carreras del 2022

```python
# Hay que crear un data frame vacia y luego ir agregando los elementos que se van creando a nuestra lista
df_results_2022=pd.DataFrame()
# Ahora corremos otra vez nuestra for a√±adiendole a la lista los valores obtenidos
for i in  range(1,23):
    url ='https://ergast.com/api/f1/2022/{}/results.json'
    response = requests.get(url.format(i))
    content = json.loads(response.content)
    df_results_2022 = df_results_2022.append(content['MRData']['RaceTable']['Races'][0]['Results'])
```

```python
df_results_2022
```

```python
# APlicamos lambda para extraer el nombre del conductor del json y solo dejarlo justo, con el nombre del conductor
df_results_2022['Driver'] = df_results_2022['Driver'].apply(lambda x: x['driverId'])
df_results_2022
```

```python
# Aplicamos lamba de nuevo para convertir los puntos a valores numericos
df_results_2022['points'] = df_results_2022['points'].apply(lambda x: float(x))
df_results_2022.head()
```

```python
# Ocupamos grouby con el parametro Driver para que se haga la suma de todos los corredores cada uno
df_results_2022.groupby(['Driver']).agg({'points':'sum'})
```

```python
# Ocupamos sort by para acomodar nuestros datos
df_results_2022.groupby(['Driver']).agg({'points':'sum'}).sort_values
```

```python
# Ascending = false para acomodarlos de manera no ascendente
df_results_2022.groupby(['Driver']).agg({'points':'sum'}).sort_values( by = 'points', ascending = False)
```

```python
# Ahora hagampos lo mismo pero para Constructor
# APlicamos lambda para extraer el nombre del conductor del json y solo dejarlo justo, con el nombre del conductor
df_results_2022['Constructor'] = df_results_2022['Constructor'].apply(lambda x: x['constructorId'])
df_results_2022
```

```python
df_results_2022.groupby(['Constructor']).agg({'points':'sum'}).sort_values( by = 'points', ascending = False)
```

## Tabla de Races

```python
content['MRData']['RaceTable']['Races']
```

```python
# Crearemos ahora una , con la informacion de las carreras
# Creamos un diccionario
races = {'season': [],
        'round': [],
        'raceName': [],
        'circuitId': [],
        'lat': [],
        'long': [],
        'country': [],
        'date': [],
        'url': [],}
```

```python
for i in list(range(2010,2023)):
    url ='https://ergast.com/api/f1/{}.json'
    response = requests.get(url.format(i))
    content = json.loads(response.content)
    
    for item in content['MRData']['RaceTable']['Races']:
    
        races['season'].append(int(item['season']))
        races['round'].append(int(item['round']))
        races['raceName'].append(item['raceName'])
        races['circuitId'].append(item['Circuit']['circuitId'])
        races['lat'].append(item['Circuit']['Location']['lat'])
        races['long'].append(item['Circuit']['Location']['long'])
        races['country'].append(item['Circuit']['Location']['country'])
        races['date'].append(item['date'])
        races['url'].append(item['url'])
```

```python
# Usamos pd.DataFrame para convertirlo en DataFrame
pd.DataFrame(races)
```
